```python
features = df[['hdlngth', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
# ÐœÐµÑ‚Ð¾Ð´ Ð»Ñ–ÐºÑ‚Ñ Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ñ— ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
wss = []  # Ð’Ð½ÑƒÑ‚Ñ€Ñ–ÑˆÐ½Ñ Ð´Ð¸ÑÐ¿ÐµÑ€ÑÑ–Ñ (WSS)
for i in range(1, 11):  # ÐŸÐµÑ€ÐµÐ²Ñ–Ñ€ÑÑ”Ð¼Ð¾ Ð´Ð»Ñ 1-10 ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wss.append(kmeans.inertia_)

# ÐŸÐ¾Ð±ÑƒÐ´Ð¾Ð²Ð° Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ° Ð»Ñ–ÐºÑ‚Ñ
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wss, marker='o', linestyle='--')
plt.title("ÐœÐµÑ‚Ð¾Ð´ Ð»Ñ–ÐºÑ‚Ñ Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²")
plt.xlabel("ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²")
plt.ylabel("WSS (Within-cluster Sum of Squares)")
plt.grid(True)
```

```python
# ÐœÐµÑ‚Ð¾Ð´ ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð² Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ñ— ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
silhouette_scores = []  # Ð¡Ð¸Ð»ÑƒÐµÑ‚Ð½Ð¸Ð¹ ÐºÐ¾ÐµÑ„Ñ–Ñ†Ñ–Ñ”Ð½Ñ‚
for i in range(2, 11):  # ÐŸÐ¾Ñ‡Ð¸Ð½Ð°Ñ”Ð¼Ð¾ Ð· 2 ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð², Ð±Ð¾ Ð´Ð»Ñ 1 Ð½ÐµÐ¼Ð¾Ð¶Ð»Ð¸Ð²Ð¾ Ð¾Ð±Ñ‡Ð¸ÑÐ»Ð¸Ñ‚Ð¸ ÑÐ¸Ð»ÑƒÐµÑ‚
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# ÐŸÐ¾Ð±ÑƒÐ´Ð¾Ð²Ð° Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ° ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð²
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('ÐœÐµÑ‚Ð¾Ð´ ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð² Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²')
plt.xlabel('ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²')
plt.ylabel('Ð¡Ð¸Ð»ÑƒÐµÑ‚Ð½Ð¸Ð¹ ÐºÐ¾ÐµÑ„Ñ–Ñ†Ñ–Ñ”Ð½Ñ‚')
plt.grid(True)
plt.show()
```

```python
# Ð’Ð¸ÐºÐ¾Ñ€Ð¸ÑÑ‚Ð°Ð½Ð½Ñ KMeans Ð´Ð»Ñ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸Ð·Ð°Ñ†Ñ–Ñ— (3 ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Ð’Ñ–Ð·ÑƒÐ°Ð»Ñ–Ð·Ð°Ñ†Ñ–Ñ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis', legend='full')

# Ð”Ð¾Ð´Ð°Ð²Ð°Ð½Ð½Ñ Ñ†ÐµÐ½Ñ‚Ñ€Ð¾Ñ—Ð´Ñ–Ð²
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')

# ÐÐ°ÑÑ‚Ñ€Ð¾ÑŽÐ²Ð°Ð½Ð½Ñ Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ°
plt.title('KMeans Clustering of Features with 3 Clusters')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
